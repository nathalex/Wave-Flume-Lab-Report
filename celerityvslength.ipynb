## Plotting theoretical relationship between celerity and wave length

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import csv
import re
from tqdm.notebook import tqdm
import math
import sympy
```

```python
data = []
g = 9.80665
h= 0.40
with open('Experimental Data-20201218/' + "wvel.csv") as csvfile:
    row_count = sum(1 for row in csvfile)
    csvfile.seek(0)
    reader = csv.reader(csvfile)
    headers = next(reader, None)
    headers = [re.sub('[\(\[].*?[\)\]]', '', header).strip().replace('  ', ' ').replace(' ', '_').lower() for header in headers]
    for row in tqdm(reader, total=row_count):
        data.append(dict(zip(headers, row)))
print(', \n'.join(data[0].keys()))
celerity = [float(d['c_/ms^-1'])/(math.sqrt(g*h)) for d in data]
length = [float(d['lmbd_/m'])/h for d in data]

plt.figure(figsize=(10,5))
plt.title("Theoretical Relationship between celerity and wavelength")
plt.ylabel("C/sqrt(gh)")
plt.xlabel("L/h")
plt.plot(length,celerity)
lengths = [3.63636/0.4, 1.481481/0.4,0.7882356/0.4]
celerities = [1.81818/(math.sqrt(g*h)),1.481481/(math.sqrt(g*h)),1.176471/(math.sqrt(g*h))]
plt.plot(lengths,celerities,"x",)
plt.grid(True, which='both')
plt.axvline(x=0,color='k')
plt.axhline(y=0, color='k')
plt.axvline(x=2,color='red')
plt.axvline(x=25,color='red')
plt.savefig("graphs/" + "celerityvslength.png")
plt.show()




```
